## Getting Started with Numpy Vectors

The numpy package has a number of powerful and fast tools for manipulating vectors. In this demo, we illustrate some of the features of the package that will be used throughout this class

For this tutorial we start by importing the package.

```python
import numpy as np
```

### Creating Vectors

We can create vectors a number of ways. We can manually create a vector by specifying the elements from a list. Not that, unlike MATLAB, there is no difference between a row and column vector.

```python
x = np.array([1, 2, 4])
print(x)
```

You can also set through a range of numbers

```python
x1 = np.arange(10)
x2 = np.arange(2, 7)
print('x1 = ' + str(x1))
print('x2 = {}'.format(x2))
```

You can also use a third step argument. The step does not need to be positive or integer.

```python
x1 = np.arange(10, 50, 5)
print(x1)

step = 2.5
x2 = np.arange(0, 10 + step/2, step)
print(x2)
```

Most operations can be vectorized meaning that the operation is applied to each component.

```python
x = np.array(range(2, 6))
y = x ** 2
print('x = {}'.format(x))
print('y = {}'.format(y))
```

### Plotting Vectors

The matplotlib package has excellent routines for plotting data and uses a very similar interface as MATLAB. To load the package use the following commands. The second line %matplotlib inline is only used when running python in jupyter notebook.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f = 3
t = np.linspace(0, 2, 100)
y = np.sin(2 * np.pi * f * t)

plt.plot(t, y)
plt.grid()
plt.xlabel('t', fontsize = 16)
plt.ylabel('sin(2pi ft)', fontsize = 16)
```

Can also stack plots

```python
ysq = y ** 2
plt.plot(t, y)
plt.plot(t, ysq)
plt.grid()
plt.xlabel('t')
```

### Indexing and Slicing Vectors

You can access individual elements by x[i]. Unlike MATLAB, the indexing starts at 0

```python
x = np.array([2, 4, 6, 8, 10])
print('Element 0 is {}'.format(x[0]))
print('Element 3 is {}'.format(x[2]))
```

Similar to MATLAB, you can obtain slices from numpy vectors, meaning subsets of components. Use the : operator as follows

```python
x = np.random.rand(10)      # 10 random elements from 0 to 1
x1 = x[2:5]
x2 = x[:4]
x3 = x[7:]
xlast = x[-1]

print('x = {}'.format(np.array_str(x, precision = 3)))
print('x[2:5] = {}'.format(np.array_str(x1, precision = 3)))
print('x[:4] = {}'.format(np.array_str(x2, precision = 3)))
print('x[7:] = {}'.format(np.array_str(x3, precision = 3)))
print('xlast = {0:5.3f}'.format(xlast))
```
